```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
df=sns.load_dataset("titanic")
df.head()
```

```python
df2=df[['survived','pclass','age','parch']]
df2.head()
```

```python
df2.isnull().sum()
```

```python
df3=df2.fillna(df2.mean())
df3
```

```python
df3.isnull().sum()
```

```python
#We have to predict survived colum
#Always put the colum u have to predict on the y axis and remaining on x
X=df3.drop("survived",axis=1)#del colum
y=df3['survived']
```

```python
#training percenet is 80% and testing 20% data it is a rule
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=51)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
#creating an object
#training and testing using StandardScaler technique
sc=StandardScaler()
sc.fit(X_train)
```

```python
sc.scale_
```

```python
X_train.describe()
```

```python
#transforming data to get mean values 0 and std 1
#returns an array
X_train_sc=sc.transform(X_train)
X_test_sc=sc.transform(X_test)
```

```python
#converting array into df
X_train_sc=pd.DataFrame(X_train_sc,columns=['pclass','age','parch'])
X_test_sc=pd.DataFrame(X_test_sc,columns=['pclass','age','parch'])
```

```python
X_train_sc.head()
```

```python
X_train_sc.describe().round(2)
```

```python
#creating an object
#training and testing using MinMaxScalar technique
mmc=MinMaxScaler()
mmc.fit(X_train)
```

```python
#transforming data to get min values 0 and max 1
#returns an array
X_train_mmc=mmc.transform(X_train)
X_test_mmc=mmc.transform(X_test)
```

```python
#converting array into df
X_train_mmc=pd.DataFrame(X_train_mmc,columns=['pclass','age','parch'])
X_test_mmc=pd.DataFrame(X_test_mmc,columns=['pclass','age','parch'])
```
