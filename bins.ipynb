```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("tips.csv")
df
```

```python
#linspace(start,end,no of values)
bins=np.linspace(min(df['total_bill']),max(df['total_bill']),4)
bins
```

```python
group_names=['low','medium','high']
```

```python
#perform binning using cut() function
df['total_bill_binned']=pd.cut(df['total_bill'],bins,labels=group_names,include_lowest=True)
```

```python
# plt.hist(df['total_bill_binned'])
df[['total_bill_binned','total_bill']].head(10)
```

```python
#see the frequency of each bin
df['total_bill_binned'].value_counts()
```

```python
# lets visualize
plt.bar(group_names,df['total_bill_binned'].value_counts())
# plt.hist(df['total_bill_binned'])
plt.show()
```
